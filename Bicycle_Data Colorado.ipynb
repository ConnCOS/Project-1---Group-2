```python
# dependencies
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
```

```python
#importing file path
#file path = "Bicycle_and_Pedestrian_Counts_in_Colorado (1).csv"
bike_data_to_load = Path("Bicycle_and_Pedestrian_Counts_in_Colorado (1).csv")
```

```python
#store as a dataframe
bike_data_to_load = pd.read_csv(bike_data_to_load)
bike_data_to_load.head()
```

```python
#Get list of data columns
list (bike_data_to_load.columns)
```

```python
Bicycle_reduced = bike_data_to_load[['stationId', 'dataCollected', 'county', 'location', 'lat',
'long',
'month',
'year',
'dataDate',
'dayOfWeek', 'hr00',
 'hr01',
 'hr02',
 'hr03',
 'hr04',
 'hr05',
 'hr06',
 'hr07',
 'hr08',
 'hr09',
 'hr10',
 'hr11',
 'hr12',
 'hr13',
 'hr14',
 'hr15',
 'hr16',
 'hr17',
 'hr18',
 'hr19',
 'hr20',
 'hr21',
 'hr22',
 'hr23',
'Location 1']]
Bicycle_reduced.head()
```

```python
#Drop all rows with missing information
Bicycle_reduced = Bicycle_reduced.dropna(how = "any")
```

```python
#Verify dropped rows
Bicycle_reduced.count()
```

```python
#Rename columns
Bicycle_reduced.rename(columns = {"stationId":"Station ID",
                                        "dataCollected" : "Data Collected",
                                        "county" : "County", 
                                        "location" : "Location",
                                        "lat" : "Lat",
                                        "long" : "Long",
                                        "month" : "Month",
                                        "year" : "Year",
                                        "dataDate" : "Data Date",
                                        "dayOfWeek" : "Day of Week",
                                        "hr00" : "Hr00", "hr01" : "Hr01", "hr02" : "Hr02",
                                        "hr03" : "Hr03", "hr04": "Hr04" , "hr05" : "Hr05", "hr06": "Hr06", "hr07" : "Hr07",
                                        "hr08" : "Hr08", "hr09" : "Hr09", "hr10" : "Hr10", "hr11" : "Hr11", "hr12" : "Hr12", "hr13": "Hr13",
                                        "hr14" : "Hr14", "hr15" : "Hr15", "hr16" : "Hr16", "hr17" : "Hr17", "hr18" : "Hr18", "hr19" : "Hr19",
                                        "hr20" : "Hr20", "hr21" : "Hr21", "hr22" : "Hr22", "hr23" : "Hr23"}, inplace = True,)
Bicycle_reduced.head()
```

```python
# Replace values within Data Collected column
Bicycle_reduced["Data Collected"] = Bicycle_reduced["Data Collected"].replace({'BIKE&PED':'BIKE & PEDESTRIAN', 'PED':'PEDESTRIAN'})
```

```python
# Verify unique values within Data colected column
Bicycle_reduced["Data Collected"].unique()
```

```python
#Group hours into single days for each county
Bicycle_reduced["Daily Count"] = Bicycle_reduced[["Hr00", "Hr01","Hr02","Hr03","Hr04","Hr05","Hr06","Hr07",
                                                 "Hr08","Hr09","Hr10","Hr11","Hr12","Hr13","Hr14","Hr15","Hr16","Hr17",
                                                 "Hr18","Hr19","Hr20","Hr21","Hr22","Hr23"]].sum(axis=1)                              
```

```python
county_total = Bicycle_reduced.groupby(["County"])["Daily Count"].sum()
county_total = county_total.sort_values(ascending=False)
county_total.head(20)
```

```python
Counties = ['Denver', 'Arapahoe', 'Boulder', 'Jefferson', 'Routt', 'El Paso', 'Adams','Larimer',
       'Summit', 'Garfield','La Plata', 'Mesa', 'Weld','Douglas', 'Eagle', 'Pueblo', 'Broomfield',
       'Montrose', 'Clear Creek','Teller']
county_total = [4544078, 940463, 798871,783944,564579, 492112 ,477310,337775,265715,244423,154463, 116729,64825,
                60982,57320,52970,17569,91,40,9]
x_axis = np.arange (len(Counties))
tick_locations = [value for value in x_axis]
plt.figure(figsize=(20,4))
plt.bar(x_axis, county_total, color='red', alpha=0.7, align="edge")
plt.xticks(tick_locations,Counties, rotation=45)
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(county_total)+100000)
plt.title("Rider Counts by County")
plt.xlabel("Counties")
plt.ylabel("Rider Daily Counts (per million)")
plt.show()
```
